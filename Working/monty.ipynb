```python
import random
l = 100000
k,s,r = 0,0,0
doors = range(3)
for i in range(l):
    c = random.choice(doors)
    g = random.choice(list(set(doors)-{c}))
    G = list(set(doors)-{c,g})[0]
    inchoice = random.choice(doors)
    opened = G if inchoice !=G else g
    switcher = list(set(doors)-{inchoice,opened})[0]
    randomer = random.choice((switcher,inchoice))
    s+=int(switcher == c)
    k+=int(inchoice == c)
    r+=int(randomer == c)
print('Switcher:',s/l)
print('Keeper:',k/l)
print('Randomer:',r/l)
```

```python
import random
l = 10000
k,s,r = 0,0,0
doors = range(100)
for i in range(l):
    c = random.choice(doors)
    g = list(set(doors)-{c})
    inchoice = random.choice(doors)
    unopened = random.choice(g) if inchoice == c else c
    switcher = unopened
    randomer = random.choice((switcher,inchoice))
    s+=int(switcher == c)
    k+=int(inchoice == c)
    r+=int(randomer == c)
print('Switcher:',s/l)
print('Keeper:',k/l)
print('Randomer:',r/l)
```
